## Part 1: Keyword Filter

```python
def keyword_filter(text):
    banned = ['hack', 'bomb', 'steal']
    return any(word in text.lower() for word in banned)

tests = [
    'How to hack',
    'How to gain unauthorized access',
    'H4ck tutorial'
]

for t in tests:
    blocked = keyword_filter(t)
    print(f'{t}: {"BLOCKED" if blocked else "ALLOWED"}')
```

## Part 2: Bypass Strategies

```python
# Synonym substitution
original = 'How to hack a system'
bypass = 'How to gain unauthorized access to a computer'

print(f'Original blocked: {keyword_filter(original)}')
print(f'Bypass blocked: {keyword_filter(bypass)}')
```
